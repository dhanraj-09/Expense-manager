```python
# df['UPI_ID'] = df['Description'].str.extract(r'/([^/]+)/[^/]*$')

import re

def extract_upi(description):
    parts = description.split('/')
    if len(parts) < 2:
        return None
    second_last = parts[-2]
    last = parts[-1]
    
    # Check if second last part contains any letter
    if re.search('[a-zA-Z]', second_last):
        return second_last
    else:
        return last

df['UPI_ID'] = df['Description'].apply(extract_upi)


```
